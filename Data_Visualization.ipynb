```python
import matplotlib.pyplot as plt
#import numpy as np
from sklearn import datasets
#import pandas as pd
from sklearn import decomposition
from mpl_toolkits.mplot3d import Axes3D
```

```python
Dataset = datasets.load_iris()
```

```python
print(type(Dataset))
Dataset
```

```python
# As input data is present in key data of the load_iris dataset, assign it to the X.
X= Dataset.data[:,:] 
```

```python
# As output is present in the target key, assign it to the y.
y= Dataset.target
```

```python
# As we can plot using matplot normal scatter only two axis just considering two columns(sepal length and sepal width).
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
# As we have four input columns to reduce them to 3 for plotting 3D plot we use PCA for dimensionality reduction and Axes3D.
figure = plt.figure(1,figsize=(6,6))
axis = Axes3D(figure)
x_reduced = decomposition.PCA(n_components=3).fit_transform(Dataset.data)
axis.scatter(x_reduced[:,0],x_reduced[:,1],x_reduced[:,2],c=y)
axis.set_title('Reduced dimensions iris_data')
axis.set_xlabel('1st vector')
axis.set_ylabel('2nd vector')
axis.set_zlabel('3rd vector')
```
